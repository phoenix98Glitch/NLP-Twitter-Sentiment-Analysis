# Creating a Pipeline to remove Punctuations, Stopwords & perform Count Vectorization

```python
# Combining last 3 steps! # Making Pipeline 
```

```python
def message_cleaning(message):
    TEST_punc_removed = [ char for char in message if char not in string.punctuation]
    TEST_punc_removed_join = "".join(TEST_punc_removed)
    TEST_punc_stopwords_removed_join_clean = [word for word in TEST_punc_removed_join.split() if word.lower() not in stopwords.words('english')]
    return TEST_punc_stopwords_removed_join_clean
```

```python
tweets_df_clean = tweets_df['tweet'].apply(message_cleaning)
```

```python
# let's compare the cleaned up and original version!
```

```python
print(tweets_df_clean[5]) # cleaned! * Blows nails *
```

```python
print(tweets_df['tweet'][5]) # original! * Ew? *
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
# Define the cleaning pipeline we defined earlier

vectorizer = CountVectorizer(analyzer = message_cleaning)
tweets_countvectorizer = vectorizer.fit_transform(tweets_df['tweet'])
```

```python
tweets_countvectorizer.shape
```

```python
X = tweets_countvectorizer
```

```python
Y = tweets_df['label'] #sentiment
```

# Train a Naive Bayes Classifier

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB

NB_classifier = MultinomialNB()
NB_classifier.fit(X_train, Y_train)
```

```python
pred1=NB_classifier.predict(X_test)
```

# Assess Trained Model Performance

```python
# Confusion Matrix #Comparing true class & predictions # TRUE+, TRUE-(type II error), FALSE+, FALSE-(type I error).
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Predicting the Test set results
Y_predict_test = NB_classifier.predict(X_test)
cm = confusion_matrix(Y_test, Y_predict_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(Y_test, Y_predict_test))
```
